```python
import numpy as np
import re

# Read the file state_matrix_step_1.txt
with open('state_matrix_step_1_channel_0.txt', 'r') as file:
    data = file.read()

# Remove non-numeric characters except for spaces, dots, and minus signs
cleaned_data = re.sub(r'[^\d\.\-\s]', '', data)

data_np = np.array(cleaned_data.split()).astype(np.float32)
data_np = data_np.reshape(84,84)
from matplotlib import pyplot as plt
# Plot the reshaped data as a black and white image
plt.imshow(data_np, cmap='gray')
plt.axis('off')  # Turn off the axis
plt.show()
```
