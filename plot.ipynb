```python
import pandas as pd

# Read the data from the CSV file row by row
df = pd.read_csv('SPS_values_opt1_grid.csv', sep=',', header=None)
```

```python
df_transposed = df.transpose()
df_transposed
```

```python
import matplotlib.pyplot as plt

# Plot the dataframe
plt.plot(df_transposed.index, df_transposed[0])
plt.xlabel('Number of Iterations')
plt.ylabel('SPS')
plt.title('Optimized Grid Values')
plt.grid(True)  # Add grid lines
# plt.xticks(range(0, 1000, 100))  # Set x-axis tick marks
plt.yticks(range(600, 1401, 100))  # Set y-axis tick marks
plt.tight_layout()  # Adjust the spacing
plt.show()
```

```python
df3 = pd.read_csv('SPS_values_opt2_grid.csv', sep=',', header=None)
df3_transposed = df3.transpose()
import matplotlib.pyplot as plt

# Plot the dataframe
plt.plot(df3_transposed.index, df3_transposed[0])
plt.xlabel('Number of Iterations')
plt.ylabel('SPS')
plt.title('Optimized_2 Grid Values')
plt.grid(True)  # Add grid lines
# plt.xticks(range(0, 1000, 100))  # Set x-axis tick marks
plt.yticks(range(600, 1401, 100))  # Set y-axis tick marks
plt.tight_layout()  # Adjust the spacing
plt.show()
```

```python
import matplotlib.pyplot as plt

# Plot the lines
plt.plot(df_transposed.index, df_transposed[0], label='Optimized_1, Precision=1')
plt.plot(df_transposed.index, df3_transposed[0], label='Optimized_2, Precision=50')
plt.plot(df2_transposed.index, df2_transposed[0], label='Non-Optimized')

# Set labels and title
plt.xlabel('Number of Iterations [Averaged across 1000 steps for each iteration]')
plt.ylabel('SPS')
plt.title('Comparison of Optimized with different precisions and Non-Optimized in Grid Observation')

# Add legend
plt.legend()

# Set y-axis limits
plt.ylim(600, 1400)

# Show the plot
plt.show()
```
